## Iniciando o SKLEARN para fazer o train test split

```python
from sklearn.model_selection import train_test_split
```

### Cirando a as varÃ¡veis X e Y - sendo X a a base sem a variÃ¡vel alvo e o Y apenas a variÃ¡vel alvo 

```python
X = energia.drop('Geracao', axis=1)
y = energia['Geracao']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
[X_train.shape, X_test.shape, y_train.shape, y_test.shape]
```

### Agora vamos importar o modelo que treinaremos com os dados

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
modelo.score(X_test, y_test)
```

### Aqui ele da o coeficiente de correlaÃ§Ã£o de cada uam das 11 variÃ¡veis independentes

```python
modelo.coef_
```

### Incluir as previsÃµes e valores reais na tabela X_test para verificar a acuracia

```python
X_test['y_pred'] = y_pred
X_test['y_test'] = y_test
```
